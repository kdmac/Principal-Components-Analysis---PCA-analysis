### Principal Component Analysis

#### Data scaling before performing PCA analysis

```python
scaler = StandardScaler()
X =  df.drop("Class",axis=1)
y = df["Class"]
```

```python
X = scaler.fit_transform(X)
```

```python
dfx = pd.DataFrame(data = X,columns=df.columns[1:])
```

```python
dfx.head()
```

```python
dfx.describe()
```

#### PCA Analysis after scaling process

```python
### PCA analysis
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=None)
```

##### Plot the explained variance ratio

```python
dfx_pac = pca.fit(dfx)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=[i+1 for i in range(len(dfx_pac.explained_variance_ratio_))],
            y=dfx_pac.explained_variance_ratio_,
           s=200, alpha=0.75,c='orange',edgecolor='k')
plt.grid(True)
plt.title("Explained variance ratio of the \nfitted principal component vector\n",fontsize=25)
plt.xlabel("Principal components",fontsize=15)
plt.xticks([i+1 for i in range(len(dfx_pac.explained_variance_ratio_))],fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel("Explained variance ratio",fontsize=15)
plt.show()
```

The above plot means that the $1^{st}$ principal component explains about 36% of the total variance in the data and the $2^{nd}$ component explians further 20%. Therefore, if we just consider first two components, they together explain 56% of the total variance or may be we can use use first 4 principal,which explain maximum variance.

```python
# Transform the scaled data set using the fitted PCA object
dfx_trans = pca.transform(dfx)
```

```python
# Put it in a data frame
dfx_trans = pd.DataFrame(dfx_trans)
```

```python
dfx_trans.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(dfx_trans[0],dfx_trans[1],c = df["Class"],edgecolors="k",alpha=0.75,s=150)
plt.grid(True)
plt.title("Class separation using first two principal components\n",fontsize=20)
plt.xlabel("Principal component-1",fontsize=15)
plt.ylabel("Principal component-2",fontsize=15)
plt.show()
```
